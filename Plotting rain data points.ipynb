```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = pd.read_csv("rainData.csv")

for i in range(5,16):
    data = data.drop(i)

data
```

```python
dates = data.columns
dates = (dates[5:]).tolist()
print(dates)
```

```python
dates.pop()
```

```python
rainfall_intensity = [[] for i in range(5)]

for i in range(5):
  for j in dates:
    rainfall_intensity[i].append(data.iloc[i][j])
```

```python
import matplotlib.pyplot as plt

plt_1 = plt.figure(figsize=(25,25))
plt.xlabel('Days')
plt.ylabel('Intensity')

plt.title('Days vs intensity graph for ', fontsize = 10)
plt.xticks(fontsize =5,rotation=45)


plt.scatter(dates, rainfall_intensity[0], color = 'g',label=data.iloc[0]['HOBLINAME'])
plt.scatter(dates, rainfall_intensity[1], color = 'r',label=data.iloc[1]['HOBLINAME'])
plt.scatter(dates, rainfall_intensity[2], color = 'b',label=data.iloc[2]['HOBLINAME'])
plt.scatter(dates, rainfall_intensity[3], color = 'k',label=data.iloc[3]['HOBLINAME'])
plt.scatter(dates, rainfall_intensity[4], color = 'y',label=data.iloc[4]['HOBLINAME'])

plt.show()
```
